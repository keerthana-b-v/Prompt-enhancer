## 📦 Cell 2: Upload, Load, and Validate Dataset

```python
import os
import pandas as pd
from google.colab import files
from datasets import Dataset

# --- Upload ---
print("📥 Please upload your 'PromptRoute_dataset_v1.csv' file:")
uploaded = files.upload()

# Accept either filename
csv_filename = None
for name in uploaded.keys():
    if name.endswith(".csv"):
        csv_filename = name
        break

if csv_filename is None:
    raise FileNotFoundError("No CSV file found in upload. Please upload PromptRoute_dataset_v1.csv")

# --- Load ---
df = pd.read_csv(csv_filename)
print(f"\n✅ Loaded: {df.shape[0]} rows, {df.shape[1]} columns")
print(f"   Columns: {list(df.columns)}")

# --- Null check ---
null_count = df['prompt'].isnull().sum() + df['label'].isnull().sum()
if null_count > 0:
    print(f"⚠️  Found {null_count} null values — dropping them")
    df = df.dropna(subset=['prompt', 'label'])
    print(f"   Rows after drop: {df.shape[0]}")
else:
    print("✅ No null values found.")

# Keep only needed columns
df = df[['prompt', 'label']].copy()
df['prompt'] = df['prompt'].astype(str).str.strip()

# --- Distribution ---
print("\n📊 Label distribution:")
dist = df['label'].value_counts()
for label, count in dist.items():
    bar = '█' * (count // 50)
    print(f"   {label:<20} {count:>5}  {bar}")

# --- Sample 3 rows per label ---
print("\n🔍 Sample prompts per label:")
for label in sorted(df['label'].unique()):
    samples = df[df['label'] == label]['prompt'].head(3).tolist()
    print(f"\n  [{label.upper()}]")
    for s in samples:
        print(f"    - {s[:80]}")

print(f"\n✅ Dataset ready. Total rows: {df.shape[0]}")
```

## 🏷️ Cell 3: Label Mapping + Model Load

Labels are detected automatically from the dataset — nothing is hardcoded.

```python
import json
from transformers import AutoModelForSequenceClassification

# --- Auto-detect labels ---
unique_labels = sorted(df['label'].unique().tolist())
num_labels = len(unique_labels)

label2id = {label: i for i, label in enumerate(unique_labels)}
id2label = {i: label for i, label in enumerate(unique_labels)}

print(f"✅ Detected {num_labels} labels (alphabetical order):")
for label, idx in label2id.items():
    print(f"   {idx}: {label}")

# --- Map string labels to integers ---
df['label'] = df['label'].map(label2id)

# --- Load model ---
print(f"\n📥 Loading distilbert-base-uncased for {num_labels} labels...")
model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id
)

total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"✅ Model loaded.")
print(f"   Total parameters:     {total_params:,}")
print(f"   Trainable parameters: {trainable_params:,}")
```

## ✂️ Cell 4: Three-Way Stratified Split + Tokenize

Split: **80% train / 10% validation / 10% test** — stratified so each label is proportionally represented in all three splits. The test set is held out completely and only evaluated at the end.

```python
from sklearn.model_selection import train_test_split
from transformers import AutoTokenizer
from datasets import Dataset

# --- Stratified 80/10/10 split ---
df_train, df_temp = train_test_split(
    df, test_size=0.20, random_state=42, stratify=df['label']
)
df_val, df_test = train_test_split(
    df_temp, test_size=0.50, random_state=42, stratify=df_temp['label']
)

print("📊 Split sizes:")
print(f"   Train:      {len(df_train):>5} rows")
print(f"   Validation: {len(df_val):>5} rows")
print(f"   Test:       {len(df_test):>5} rows")

print("\n📊 Label distribution per split:")
for split_name, split_df in [("Train", df_train), ("Val", df_val), ("Test", df_test)]:
    counts = split_df['label'].value_counts().sort_index()
    labels_str = ", ".join([f"{id2label[k]}:{v}" for k, v in counts.items()])
    print(f"   {split_name}: {labels_str}")

# --- Convert to HuggingFace Dataset ---
train_dataset = Dataset.from_pandas(df_train.reset_index(drop=True))
val_dataset   = Dataset.from_pandas(df_val.reset_index(drop=True))
test_dataset  = Dataset.from_pandas(df_test.reset_index(drop=True))

# --- Tokenize ---
print("\n✂️  Loading tokenizer and tokenizing...")
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def tokenize_function(examples):
    return tokenizer(
        examples['prompt'],
        max_length=128,
        truncation=True,
        padding='max_length'
    )

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset   = val_dataset.map(tokenize_function, batched=True)
test_dataset  = test_dataset.map(tokenize_function, batched=True)

# Set format for PyTorch
cols = ['input_ids', 'attention_mask', 'label']
train_dataset = train_dataset.remove_columns([c for c in train_dataset.column_names if c not in cols])
val_dataset   = val_dataset.remove_columns([c for c in val_dataset.column_names if c not in cols])
test_dataset  = test_dataset.remove_columns([c for c in test_dataset.column_names if c not in cols])

train_dataset.set_format('torch')
val_dataset.set_format('torch')
test_dataset.set_format('torch')

print("✅ Tokenization complete.")
print(f"   Columns: {train_dataset.column_names}")
```

## 📈 Cell 5: Define Evaluation Metrics

```python
import numpy as np
import evaluate
from sklearn.metrics import classification_report

accuracy_metric = evaluate.load('accuracy')
f1_metric       = evaluate.load('f1')

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    acc = accuracy_metric.compute(predictions=predictions, references=labels)['accuracy']
    f1  = f1_metric.compute(predictions=predictions, references=labels, average='weighted')['f1']

    return {'accuracy': acc, 'f1': f1}

print("✅ Metrics defined: accuracy + weighted F1")
print("   Per-class breakdown will be shown after final test evaluation.")
```

## ⚙️ Cell 6: Training Arguments

Key settings: **fp16=True** for ~40% faster training on T4, **warmup_ratio** instead of warmup_steps, **batch_size 32**, **lr 3e-5**, **6 epochs**.

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./PromptRoute-classifier',
    num_train_epochs=6,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=64,
    warmup_ratio=0.1,
    weight_decay=0.01,
    learning_rate=3e-5,
    eval_strategy='epoch',
    save_strategy='epoch',
    load_best_model_at_end=True,
    metric_for_best_model='f1',
    greater_is_better=True,
    logging_steps=25,
    fp16=True,
    dataloader_num_workers=2,
    report_to='none',
)

print("✅ TrainingArguments configured:")
print(f"   Epochs:         {training_args.num_train_epochs}")
print(f"   Batch size:     {training_args.per_device_train_batch_size}")
print(f"   Learning rate:  {training_args.learning_rate}")
print(f"   fp16:           {training_args.fp16}")
print(f"   Warmup ratio:   {training_args.warmup_ratio}")
print(f"   Best model by:  {training_args.metric_for_best_model}")
```

## 🔄 Cell 7: Checkpoint Resume Check

If your Colab session disconnected mid-training, this cell finds the last saved checkpoint so training resumes from there instead of starting over.

```python
import os

checkpoint_dir = './PromptRoute-classifier'
resume_path = None

if os.path.exists(checkpoint_dir):
    checkpoints = sorted([
        d for d in os.listdir(checkpoint_dir)
        if d.startswith('checkpoint')
    ])
    if checkpoints:
        latest = checkpoints[-1]
        resume_path = os.path.join(checkpoint_dir, latest)
        print(f"✅ Found checkpoint: {resume_path}")
        print("   Training will RESUME from this checkpoint.")
    else:
        print("ℹ️  No checkpoint found. Starting fresh training.")
else:
    print("ℹ️  No checkpoint directory found. Starting fresh training.")
```

## 🚀 Cell 8: Train the Model

```python
import time
from transformers import Trainer

# --- Estimate training time ---
epochs     = int(training_args.num_train_epochs)
batch_size = training_args.per_device_train_batch_size
total_rows = len(train_dataset)
est_secs   = (total_rows / batch_size) * epochs * 0.15
est_mins   = max(1, round(est_secs / 60))
print(f"⏱️  Estimated training time: ~{est_mins} minutes on T4 GPU")
print(f"   (Train rows: {total_rows}, Batch: {batch_size}, Epochs: {epochs})\n")

# --- Build Trainer ---
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,       # validation set only — test set is held out
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

# --- Train ---
print("🔥 Starting training...")
start_time = time.time()
trainer.train(resume_from_checkpoint=resume_path)
elapsed = (time.time() - start_time) / 60

print(f"\n✅ Training complete in {elapsed:.1f} minutes.")
```

## 🧪 Cell 9: Final Evaluation on Held-Out Test Set

This is the only time the test set is evaluated. Shows overall metrics AND per-class breakdown so you can see which labels are weak.

```python
import numpy as np
from sklearn.metrics import classification_report

# --- Overall metrics on test set ---
print("📊 Evaluating on held-out test set...\n")
test_results = trainer.evaluate(eval_dataset=test_dataset)
print(f"🏆 Test Accuracy:    {test_results['eval_accuracy']:.4f}")
print(f"🏆 Test F1 Weighted: {test_results['eval_f1']:.4f}")

# --- Per-class classification report ---
print("\n📋 Per-class breakdown:")
predictions_output = trainer.predict(test_dataset)
y_pred = np.argmax(predictions_output.predictions, axis=-1)
y_true = predictions_output.label_ids

target_names = [id2label[i] for i in range(num_labels)]
report = classification_report(y_true, y_pred, target_names=target_names, digits=3)
print(report)

# Highlight weak labels (f1 < 0.80)
print("⚠️  Labels with F1 < 0.80 (need more training data):")
from sklearn.metrics import f1_score
per_class_f1 = f1_score(y_true, y_pred, average=None)
weak_found = False
for i, f1_val in enumerate(per_class_f1):
    if f1_val < 0.80:
        print(f"   ❌ {id2label[i]:<20} F1: {f1_val:.3f}")
        weak_found = True
if not weak_found:
    print("   ✅ All labels above 0.80 F1 — good model quality!")
```

## ⚡ Cell 12: Export to ONNX via Native PyTorch Compiler

```python
import os
import sys

# Auto-install PyTorch ONNX compiler dependencies in 5 seconds
print("📦 Installing ONNX compiler dependencies...")
!pip install -q onnx onnxscript

import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

model_dir = "/content/PromptRoute-classifier-final"
onnx_dir = "/content/PromptRoute-onnx"
os.makedirs(onnx_dir, exist_ok=True)

print("📥 Loading PyTorch model and tokenizer...")
model = AutoModelForSequenceClassification.from_pretrained(model_dir)
tokenizer = AutoTokenizer.from_pretrained(model_dir)

# Set model to evaluation mode for inference
model.eval()

# Create dummy inputs matching the shape expected by DistilBERT (batch_size=1, sequence_length=128)
dummy_input_ids = torch.ones(1, 128, dtype=torch.long)
dummy_attention_mask = torch.ones(1, 128, dtype=torch.long)
dummy_inputs = (dummy_input_ids, dummy_attention_mask)

print("⚡ Compiling model to ONNX using native PyTorch exporter...")
# Export model directly with fully declared dynamic axes
torch.onnx.export(
    model,
    dummy_inputs,
    os.path.join(onnx_dir, "model.onnx"),
    input_names=["input_ids", "attention_mask"],
    output_names=["logits"],
    dynamic_axes={
        "input_ids": {0: "batch_size", 1: "sequence_length"},
        "attention_mask": {0: "batch_size", 1: "sequence_length"},
        "logits": {0: "batch_size", 1: "num_labels"}
    },
    opset_version=14
)

print("✅ Native PyTorch ONNX export complete! Saved to:", onnx_dir)
```

## 🗜️ Cell 13: Dynamic INT8 Quantization

Compresses model from ~268MB to under 70MB using signed INT8 (QInt8). Prints size comparison before and after.

```python
import os
import json
import shutil
from onnxruntime.quantization import quantize_dynamic, QuantType

model_fp32  = '/content/PromptRoute-onnx/model.onnx'
model_quant = '/content/PromptRoute-onnx/model_quantized.onnx'

size_before = os.path.getsize(model_fp32) / (1024**2)
print(f"📏 Original model size: {size_before:.1f} MB")
print("⚖️  Applying dynamic INT8 quantization (QInt8)...")

# QInt8 (signed) is correct for transformer models
quantize_dynamic(
    model_input=model_fp32,
    model_output=model_quant,
    weight_type=QuantType.QInt8
)

size_after = os.path.getsize(model_quant) / (1024**2)
reduction  = (1 - size_after / size_before) * 100

print(f"✅ Quantization complete!")
print(f"   Original:   {size_before:.1f} MB")
print(f"   Quantized:  {size_after:.1f} MB")
print(f"   Reduction:  {reduction:.1f}%")

if size_after > 70:
    print(f"⚠️  Warning: Quantized model is {size_after:.1f}MB — above 70MB target.")
    print("   The extension will still work but first-load will be slower.")
else:
    print(f"✅ Size is under 70MB target — good for Chrome extension bundling.")

# Copy label_map.json to onnx folder
shutil.copy(
    '/content/PromptRoute-classifier-final/label_map.json',
    '/content/PromptRoute-onnx/label_map.json'
)
print("✅ label_map.json copied to ONNX folder.")
```

## 🧪 Cell 14: Test Quantized Model — 12 Prompts

Runs all 12 test prompts through the quantized ONNX model and checks each prediction against the expected label.

```python
import json
import numpy as np
import onnxruntime as ort
from transformers import AutoTokenizer

# Load tokenizer and ONNX session
tokenizer_test = AutoTokenizer.from_pretrained('/content/PromptRoute-classifier-final')
session = ort.InferenceSession('/content/PromptRoute-onnx/model_quantized.onnx')

with open('/content/PromptRoute-onnx/label_map.json', 'r') as f:
    lmap = json.load(f)
id2label_test = lmap['id2label']

# 12 test prompts with expected labels
test_cases = [
    ("write a short story about a lost dog",                           "creative"),
    ("debug this javascript function that returns undefined",          "code"),
    ("explain how black holes form",                                   "factual"),
    ("create a 30 day fitness plan",                                   "planning"),
    ("analyze the pros and cons of remote work",                       "analysis"),
    ("give me the output in json format with name and age fields",     "structured_output"),
    ("search the web for latest news on openai",                       "agentic"),
    ("solve for x when 5x minus 3 equals 22",                          "math"),
    ("hi",                                                             "skip"),
    ("ok thanks",                                                      "skip"),
    ("write a comprehensive guide to machine learning",                "longform"),
    ("compare python and javascript for web development",              "analysis"),
]

def softmax(x):
    e_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return e_x / e_x.sum(axis=-1, keepdims=True)

def predict_single(text):
    inputs = tokenizer_test(
        text, return_tensors='np',
        max_length=128, truncation=True, padding='max_length'
    )
    ort_inputs = {
        'input_ids':      inputs['input_ids'].astype(np.int64),
        'attention_mask': inputs['attention_mask'].astype(np.int64)
    }
    logits = session.run(None, ort_inputs)[0]
    probs  = softmax(logits)[0]
    pred_id = int(np.argmax(probs))
    return id2label_test[str(pred_id)], float(probs[pred_id])

print("🧪 Testing quantized model on 12 prompts")
print("=" * 80)

passed = 0
for prompt, expected in test_cases:
    predicted, confidence = predict_single(prompt)
    match = predicted == expected
    status = "✅ PASS" if match else "❌ FAIL"
    if match:
        passed += 1
    print(f"{status} | Expected: {expected:<18} Predicted: {predicted:<18} ({confidence:.1%})")
    print(f"       Prompt: '{prompt[:65]}'")
    print("-" * 80)

print(f"\n🏆 Result: {passed}/{len(test_cases)} prompts correctly classified")
if passed < 9:
    print("⚠️  Less than 9/12 correct — consider adding more training data for weak labels.")
else:
    print("✅ Model quality looks good for production.")
```
